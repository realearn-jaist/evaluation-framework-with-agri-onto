```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import pandas as pd
```

```python
data=pd.read_excel('/content/Heatmap cosine.xlsx')
```

```python
embedding_vectors=data
```

```python
cosine_similarities = cosine_similarity(embedding_vectors)
```

```python
# Create a heatmap of the cosine similarities
plt.figure(figsize=(10, 8))
sns.heatmap(cosine_similarities, cmap="YlGnBu", xticklabels=False, yticklabels=False)
plt.title("Cosine Similarity Heatmap of Embedding Vectors")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE


data = pd.read_excel('/content/Heatmap.xlsx')
labels = data.iloc[:, 0].tolist()

# Extract the embeddings as a numpy array
embedded_vectors = data.iloc[:, 1:].values

# Initialize t-SNE with desired hyperparameters
tsne = TSNE(n_components=2, random_state=42)

# Perform t-SNE on the embedded vectors
embedded_tsne = tsne.fit_transform(embedded_vectors)

# Create a scatter plot to visualize the t-SNE results
plt.figure(figsize=(10, 8))
plt.scatter(embedded_tsne[:, 0], embedded_tsne[:, 1], alpha=0.7)

 #Add labels to the points if available

for i, label in enumerate(labels):
    plt.text(embedded_tsne[i, 0], embedded_tsne[i, 1], str(label), fontsize=6)

plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE Visualization')
plt.show()
```
